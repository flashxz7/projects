```python
#small wordle game cheater

import random

print("Welcome to Wordle!")

with open("wordle_list.txt", "r") as file:
    words = [word.strip().lower() for word in file if len(word.strip()) == 5]


secret_word = random.choice(words)

print("\nThe computer has selected a secret five-letter English word.")
print("You have six guesses to determine it.")
print("Use UPPERCASE for correct letters in the correct position (green).")
print("Use lowercase for correct letters in the wrong position (yellow).")
print("'-' means the letter is not in the word at all (gray).")
print("Good luck!\n")


green_letters = [''] * 5        
yellow_letters = []             
gray_letters = set()      

for attempt in range(6):
    guess = input(f"Guess #{attempt+1}: ").strip().lower()

    if len(guess) != 5:
        print("Your guess must be exactly 5 letters.\n")
        continue

    if guess not in words:
        print("That word is not in the dictionary.\n")
        continue

    result = ""
    for i in range(5):
        if guess[i] == secret_word[i]:
            result += guess[i].upper()
            green_letters[i] = guess[i]
        elif guess[i] in secret_word:
            result += guess[i].lower()
            yellow_letters.append((guess[i], i))
        else:
            result += "-"
            gray_letters.add(guess[i])

    print("Status:", result)

    
    if guess == secret_word:
        print(f"\n Correct! You guessed the word in {attempt+1} tries!")
        break

    
    possible_words = []
    for word in words:
        match = True

        
        for i in range(5):
            if green_letters[i] and word[i] != green_letters[i]:
                match = False
                break

        
        for (yl, pos) in yellow_letters:
            if yl not in word or word[pos] == yl:
                match = False
                break

        
        for gl in gray_letters:
            
            if gl in word and gl not in green_letters and all(yl[0] != gl for yl in yellow_letters):
                match = False
                break

        if match:
            possible_words.append(word)

    print(f"Possible matches ({len(possible_words)}):", ", ".join(possible_words[:10]) + "\n")

else:
    print("\nOut of guesses!")
    print("The word was:", secret_word.upper())
```
