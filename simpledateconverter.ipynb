```python
import re
import requests
from bs4 import BeautifulSoup
from nltk import sent_tokenize
from datetime import datetime


date_pattern = re.compile(r'\b(?:January|February|March|April|May|June|July|August|September|October|November|December)\s+\d{1,2},\s+\d{4}')


def convert(date_string):
    try:
        date_obj = datetime.strptime(date_string, "%B %d, %Y")
        return date_obj.strftime("%Y-%m-%d")
    except:
        return "Could not parse"

url = 'https://en.wikipedia.org/wiki/American_Revolution'
response = requests.get(url)
source = response.text
soup = BeautifulSoup(source,'html.parser')
list = soup.find_all('p')

sentence_number = 0

for p in list:
    sent = sent_tokenize(p.text)
    for se in sent:
        dates = date_pattern.findall(se)
        if dates:
            sentence_number += 1
            print(f"Sentence #{sentence_number}")
            print('-'*100)
            print(se)
            for d in dates:
                print(f"{d} -> {convert(d)}")
```
